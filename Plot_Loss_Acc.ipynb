```python
import matplotlib.pyplot as plt
```

```python
def plot_all():
    loss_train = []
    loss_val = []
    acc_train = []
    acc_val = []

    with open("Logs_attention.txt") as f:
        lines = f.readlines()
    for line in lines:
        line = line.split()
        loss_train.append(float(line[7]))
        loss_val.append(float(line[9]))
        acc_train.append(float(line[16]))
        acc_val.append(float(line[18]))
        
    plt.figure(figsize=(20,5))
    plt.plot(loss_train, label='Training Loss')
    plt.plot(loss_val, label='Validaotin Loss')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()
    
    plt.figure(figsize=(20,5))
    plt.plot(acc_train, label='Training Accuracy')
    plt.plot(acc_val, label='Validaotin Accuracy')
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()
```

```python
plot_all()
```
